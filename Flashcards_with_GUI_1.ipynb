```python
from tkinter import *
import pandas as pd
from random import randint

root = Tk()
root.title("opencampus.sh - Flashcard")
root.geometry("550x500")

words = pd.read_csv("https://raw.githubusercontent.com/kboroz/Python_Programming_Project/main/Words.csv")
eng_word = words.English.tolist()
ger_word = words.German.tolist()
# get a count of words
count = len(eng_word)

def next():
    global hinter, hint_count
    #clear the screen
    answer_label.config(text="")
    my_entry.delete(0, END)
    hint_label.config(text="")
    #reset hints
    hinter = ""
    hint_count = 0

    #creat random selection
    global random_word
    random_word = randint(0, count-1)
    #update label with front word
    question_label.config(text=eng_word[random_word])

def answer():
    if my_entry.get().title() == ger_word[random_word]:
        answer_label.config(text=f"Correct! {eng_word[random_word]} is {ger_word[random_word]}.")
    else:
        answer_label.config(text=f"Wrong! {eng_word[random_word]} is not {my_entry.get().title()}.")


#keep track of hints
hinter = ""
hint_count = 0

def hint():
    global hinter
    global hint_count
    if hint_count < len(ger_word[random_word]):
        hinter += ger_word[random_word][hint_count]
        hint_label.config(text=hinter)
        hint_count +=1


question_label = Label(root, text="", font=("Helvetica", 38))
question_label.pack(pady=50)

answer_label = Label(root, text="")
answer_label.pack(pady=30)

my_entry = Entry(root, font=("Helvetica", 18))
my_entry.pack(pady=20)

# create buttons
button_frame = Frame(root)
button_frame.pack(pady=20)

answer_button = Button(button_frame, text="Answer", command=answer)
answer_button.grid(row=0, column=0, padx=20)

next_button = Button(button_frame, text="Next", command=next)
next_button.grid(row=0, column=1)

hint_button = Button(button_frame, text="Hint", command=hint)
hint_button.grid(row=0, column=2, padx=20)

# create hint label
hint_label = Label(root, text="", font=("Helvetica", 16))
hint_label.pack(pady=20)

# run next function when program starts
next()

root.mainloop()
```
