```python
class Flashcard:
    def __init__(self, front, back):
        self.front = front
        self.back = back

def add_card(cards):
    front = input("Enter the front of the card: ")
    back = input("Enter the back of the card: ")
    cards.append(Flashcard(front, back))
    return cards

def edit_card(cards):
    view_cards(cards)
    index = int(input("Enter the card number to edit: ")) - 1
    if 0 <= index < len(cards):
        cards[index].front = input("Enter the new front of the card: ")
        cards[index].back = input("Enter the new back of the card: ")
    else:
        print("Invalid card number")
    return cards

def delete_card(cards):
    view_cards(cards)
    index = int(input("Enter the card number to delete: ")) - 1
    if 0 <= index < len(cards):
        cards.pop(index)
    else:
        print("Invalid card number")
    return cards

def view_cards(cards):
    for i, card in enumerate(cards):
        print(f"{i+1}. Front: {card.front}, Back: {card.back}")

def quiz(cards):
    import random
    count = int(input("How many cards do you want to study? "))
    random.shuffle(cards)
    correct_answers = 0
    for i in range(min(count, len(cards))):
        print(f"Card {i+1}: {cards[i].front}")
        answer = input("Enter your answer: ")
        if answer == cards[i].back:
            correct_answers += 1
    print(f"Quiz finished. You got {correct_answers} out of {min(count, len(cards))} correct.")

def main():
    cards = []
    while True:
        print("Select an option:")
        print("1. Start quiz")
        print("2. Add card")
        print("3. Edit card")
        print("4. Delete card")
        print("5. View cards")
        choice = input("Enter choice (1/2/3/4/5): ")

        if choice == '1':
            quiz(cards)
        elif choice == '2':
            cards = add_card(cards)
        elif choice == '3':
            cards = edit_card(cards)
        elif choice == '4':
            cards = delete_card(cards)
        elif choice == '5':
            view_cards(cards)
        else:
            print("Invalid input")

if __name__ == "__main__":
    main()
```
