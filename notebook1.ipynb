```python
import pandas as pd




cards_df=pd.read_excel("words.xlsx")

cards_df["front"] = cards_df["front"].str.lower()
cards_df["back"] = cards_df["back"].str.lower()
cards_df.head(2)

front_list = cards_df["front"].to_list()
back_list = cards_df["back"].to_list()

import random
def start_quiz():
  counter= int(input("How many cards do you want to study  "))
  result= 0
  random_cards_index= random.sample(range(0,len(front_list)), counter)
  # display_cards=
  for i in random_cards_index:
    response= input(f"Enter the answer for {front_list[i]} -->")
    if response== back_list[i]:
      result+=1
    else:
      pass
  print(f"You scored {result} out of {counter}")

def add_cards():
  while True:
    choice= input(f"do you want to add a card? type y for yes and n for no  ")
    choice=choice.lower()

    if choice== "y":
      card_front= input("write in the front of the card--->  ")
      card_back= input("write in the back of the card--->   ")


      front_list.append(card_front)
      back_list.append(card_back)

      print("you must enter both front and back")

    elif choice== "n":
      break
    else:
      print("invalid input")

  save_to_excel()

def delete_cards():
  while True:
    choice= input("do you want to delete a card? type y for yes or n for no  ")
    choice=choice.lower()

    if choice== 'y':
      card_to_delete= input("type in the card you want to delete  ")
      try:
        card_to_delete_index= front_list.index(card_to_delete)
        front_list.pop(card_to_delete_index)
        back_list.pop(card_to_delete_index)
      except ValueError:
        try:
          card_to_delete_index= back_list.index(card_to_delete)
          front_list.pop(card_to_delete_index)
          back_list.pop(card_to_delete_index)

        except ValueError:
          print("This card does not exist")


    elif choice == 'n':
      break
    else:
      print("invalid input")

  save_to_excel()

def edit_cards():
  while True:
    choice= input("do you want to edit a card? type y for yes or n for no  ")
    choice=choice.lower()

    if choice== 'y':
      card_to_edit= input("type in the card you want to edit ")
      try:
        card_to_edit_index= front_list.index(card_to_edit)
        front_list[card_to_edit_index] = input((f"enter the replacement text "))
      except ValueError:
        try:
          card_to_edit_index= front_list.index(card_to_edit)
          back_list[card_to_edit_index] = input((f"enter the replacement tex t"))

        except ValueError:
          print("This card does not exist")

    elif choice == 'n':
      break
    else:
      print("invalid input")
  save_to_excel()

def view_result():
  card_table= pd.DataFrame({"front": front_list,"back": back_list})
  pd.set_option('display.max_rows', None)
  print(card_table)
  pd.reset_option('display.max_rows')

def save_to_excel():
  cards_df = pd.DataFrame({'front':front_list, 'back':back_list})
  cards_df.to_excel("words.xlsx", index= False)

def main_program():
  while True:
    user_input= int(input("What do you want to do? Choose 1=start quiz, 2=add, 3=edit, 4=delete, 5=view, 6=exit: "))
    if user_input == 1:
      start_quiz()
    elif user_input == 2:
      add_cards()
    elif user_input==3:
      edit_cards()
    elif user_input==4:
      delete_cards()
    elif user_input==5:
      view_result()
    elif user_input==6:
      print("bye")
      break
    else:
      print("invalid input")

main_program()


```
